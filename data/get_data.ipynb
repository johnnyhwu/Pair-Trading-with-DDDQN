## List of Stock Name

```python
stock_name = [
    "AAPL",
    "MSFT",
    "GOOG",
    "AMZN",
    "GE",

    "FB",
    "GOOGL",
    "TSLA",
    "TSM",
    "BABA",
    
    "NVDA",
    "V",
    "WMT",
    "MA",
    "ASML",
    
    "DIS",
    "PYPL",
    "ADBE",
    "CMCSA",
    "ORCL",

    "CSCO",
    "VZ",
    "NFLX",
    "CRM",
    "INTC",

    "DHR",
    "ACN",
    "T",
    "AVGO",
    "TMUS"
]
```

## Download Historical Stock Price Data

```python
for stock in stock_name:
    data = yf.Ticker(stock)
    time.sleep(0.5)
    data = data.history(period="max")

    if data.shape[0] == 0:
        print("{} not found".format(stock))
        time.sleep(0.5)
        continue
    
    data.to_csv(path_or_buf="/gdrive/MyDrive/專題/data/{}.csv".format(stock))
    print("{} downloaded".format(stock))
    time.sleep(0.5)
```
